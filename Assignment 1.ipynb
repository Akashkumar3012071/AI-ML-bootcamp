```python
#Methods of Strings
S="Kharagpur"


print(S.count("a"))
print(S.find("g"))
print(S.casefold())
print(S.capitalize())
print(S.islower())
```

```python
#Methods of List
L=[89,56,52,34,56,78,989,232,43,56]
L.append(23)
print(L)
print(L.count(56))
L.insert(5,33)
print(L)
L.remove(78)
print(L)
L.sort()
print(L)
```

```python
#methods of Tuples
T=("egg","bread","butter","music","jam","egg")
print(T)
print(type(T))
print(len(T))
print(T[3])
print(T.count("egg"))
print(T.index("jam"))

```

```python
#methods of set

X={1,3,651,75,3445,345,45,2,3,3,3,3}
Y={3,75,45,234,54,345,323,898}
print(X)
X.add(90)
print(X)
Z=set.difference(X,Y)
print(Z)
Z=set.intersection(X,Y)
print(Z)
Z=set.union(X,Y)
print(Z)
```

```python
#method of dictionaries

D = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}

print(D.items())
print(D.keys())
print(D.get("brand"))
print(D.fromkeys("year"))
print(D.values())
```
